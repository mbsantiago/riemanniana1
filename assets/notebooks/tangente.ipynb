```python
import ipyvolume.pylab as p3
import numpy as np
import sympy as sp
```

```python
from sympy.abc import x,y,z
```

```python
f_sym = sp.sin(x)+3
g_sym = sp.cos(x)*0.2+0.5
```

```python
f = sp.lambdify(x, f_sym, "numpy")
g = sp.lambdify(x, g_sym, "numpy")
```

```python
mapx = f_sym*sp.cos(y)
mapy = f_sym*sp.sin(y)
mapz = g_sym
mapxn = sp.lambdify((x,y), mapx, "numpy")
mapyn = sp.lambdify((x,y), mapy, "numpy")
```

```python
parx_x = sp.diff(mapx, x)
pary_x = sp.diff(mapy, x)
parz_x = sp.diff(mapz, x)
parx_y = sp.diff(mapx, y)
pary_y = sp.diff(mapy, y)
parz_y = sp.diff(mapz, y)
px_x = sp.lambdify((x,y), parx_x, "numpy")
py_x = sp.lambdify((x,y), pary_x, "numpy")
pz_x = sp.lambdify((x,y), parz_x, "numpy")
px_y = sp.lambdify((x,y), parx_y, "numpy")
py_y = sp.lambdify((x,y), pary_y, "numpy")
pz_y = sp.lambdify((x,y), parz_y, "numpy")
```

```python
tp = np.array([.7,2.5])
```

```python
time = np.linspace(0, np.pi*5, 150)
```

```python
tpoints = [[1.0+2*np.cos(t*0.2), 2.0+t] for t in time]
```

```python
Vxx = px_x(tp[0],tp[1])
Vxy = py_x(tp[0],tp[1])
Vxz = pz_x(tp[0],tp[1])
Vyx = px_y(tp[0],tp[1])
Vyy = py_y(tp[0],tp[1])
Vyz = pz_y(tp[0],tp[1])
```

```python
pt=[mapxn(tp[0],tp[1]),mapyn(tp[0],tp[1]),g(tp[0])]
```

```python
pt
```

```python
a = np.arange(-4, 4, 0.1)
U, V = np.meshgrid(a, a)
X = f(V)*np.cos(U)
Y = f(V)*np.sin(U)
Z = g(V)
a = np.arange(-1, 1, 0.05)
U, V = np.meshgrid(a, a)
Xt = Vxx*U+Vyx*V+np.float(pt[0])
Xt = [px_x(p[0],p[1])*U+px_y(p[0],p[1])*V+mapxn(p[0],p[1]) for p in tpoints]
Yt = [py_x(p[0],p[1])*U+py_y(p[0],p[1])*V+mapyn(p[0],p[1]) for p in tpoints]
Zt = [pz_x(p[0],p[1])*U+pz_y(p[0],p[1])*V+g(p[0]) for p in tpoints]


p3.figure()
s = p3.plot_mesh(Xt, Zt, Yt)
p3.plot_surface(X, Z, Y, color="orange")
p3.animation_control(s, interval=500)
p3.show()
```
